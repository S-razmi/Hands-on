## **Introduction to Non-Parametric Tests:**

<a id="what_is"></a>

### **What are Non-Parametric Tests:**

<div align='left'  style="text-align: left">
<p>Non-parametric tests are statistical methods used for analyzing data when the assumptions of normality or homogeneity of variance are not met. Unlike parametric tests, non-parametric tests make fewer assumptions about the underlying population distribution, making them robust and versatile in various situations.</p>


**Benefits:**
- **Robustness:** Non-parametric tests are less sensitive to outliers and do not require strict adherence to distributional assumptions.
- **Applicability:** They are suitable for both continuous and ordinal data, offering flexibility in diverse research settings.
- **Simplicity:** Non-parametric tests are often simpler to compute and interpret, making them accessible to a wider audience.

**Drawbacks:**
- **Less Power:** Non-parametric tests may have less statistical power compared to their parametric counterparts, especially with larger sample sizes.
- **Loss of Information:** They may lose some information by focusing on ranks or categories rather than actual measurement values.
- **Limited Application:** In certain cases, when parametric assumptions are reasonably met, non-parametric tests might be less efficient.

Understanding the benefits and drawbacks of non-parametric tests allows researchers to choose the most appropriate method for their specific analysis needs.
### **When to Use Non-Parametric Tests**

Non-parametric tests are particularly suitable in several scenarios where the assumptions of traditional parametric tests are not met or when dealing with specific types of data:

1. **Non-Normal Distribution:**
   - Use non-parametric tests when the data does not follow a normal distribution, as non-parametric tests are less sensitive to deviations from normality.

2. **Ordinal or Categorical Data:**
   - Opt for non-parametric tests when dealing with ordinal or categorical data that lacks a clear numerical interpretation, as parametric tests may require interval or ratio data.

3. **Skewed or Outlier-Prone Data:**
   - When the dataset contains skewed distributions or potential outliers, non-parametric tests are more robust and less influenced by extreme values.

4. **Small Sample Sizes:**
   - Non-parametric tests can be preferable with small sample sizes, as they do not rely on large sample assumptions that parametric tests often require for accurate results.

5. **Equality of Variances Assumption Not Met:**
   - If the assumption of equal variances is violated, non-parametric tests can be used instead of their parametric counterparts, which may be sensitive to unequal variances.

6. **Nominal Data:**
   - When dealing with nominal data (categories with no inherent order), non-parametric tests like the Chi-Square test are appropriate.

7. **Transformation Not Feasible:**
   - In situations where data transformation to meet parametric assumptions is not feasible or would lead to loss of meaningful information, non-parametric tests are a viable alternative.

By recognizing these scenarios, researchers can make informed decisions on when to employ non-parametric tests to ensure the validity and reliability of their statistical analyses.

<a id="mann"></a>

### **Mann-Whitney U Test:**

<div align='justify'>
<p>

The Mann-Whitney U Test is a non-parametric test used to compare the distributions of two independent samples. It is particularly useful when dealing with non-normally distributed data or ordinal data. The test assesses whether the probability of an observation in one sample is equal to the probability of an observation in the other sample.

**Counterpart Parametric Test:**
- **Independent Samples t-Test:**
  - The Mann-Whitney U Test is the non-parametric equivalent of the independent samples t-test. It is employed when assumptions of normality are not met, making it suitable for a wide range of data types.

**Manual Execution:**
1. **Ranking Data:** Combine data from both samples, rank all observations, and assign average ranks to tied values.
2. **Calculate U-Statistic:** Compute the U statistic, which represents the sum of ranks for one sample, adjusting for ties.
3. **Compare U to Critical Value:** Compare the U statistic to critical values from the Mann-Whitney U distribution or use a significance table.
4. **Determine Significance:** If the calculated U falls below the critical value, the difference between the samples is considered statistically significant.
</pr>

<a id="signed"></a>

### **Signed Rank Test:**

<div align='justify'>

The Wilcoxon Signed-Rank Test is a non-parametric test used to compare two related samples, typically paired observations. It assesses whether the distribution of the differences between paired observations is symmetrically centered around zero. This test is suitable when the assumptions of the parametric paired samples t-test are not met, especially in the case of non-normally distributed data.

**Counterpart Parametric Test:**
- **Paired Samples t-Test:**
  - The Wilcoxon Signed-Rank Test is the non-parametric equivalent of the paired samples t-test. It is used when the distribution of differences is not assumed to be normal.

**Manual Execution:**
1. **Calculate Differences:** Find the differences between paired observations.
2. **Rank Absolute Differences:** Rank the absolute values of the differences, assigning the average rank to tied values.
3. **Calculate the Test Statistic (W):** Sum the ranks of positive or negative differences (ignoring zeros).
4. **Compare W to Critical Value:** Use tables or software to find the critical value for W.
5. **Determine Significance:** If the calculated W falls below (or above) the critical value, reject the null hypothesis.

Performing the Wilcoxon Signed-Rank Test manually involves ranking the absolute differences and comparing the test statistic to critical values. Adjust the sample data or significance level as needed for your specific analysis.
</div>

<a id="kruskall"></a>

### **Kruskall-Wallis Test:**

<div align='justify'>

The Kruskal-Wallis test is a non-parametric test used to determine if there are statistically significant differences among three or more independent groups. It's an extension of the Mann-Whitney U test for more than two groups. The test compares the ranks of observations from different groups to assess whether their distributions are similar or if there is a significant difference.

**Counterpart Parametric Test:**
- **One-Way ANOVA:**
  - The Kruskal-Wallis test is the non-parametric equivalent of the one-way analysis of variance (ANOVA). It is used when the assumptions of ANOVA, such as normality and homogeneity of variances, are not met.

**Manual Execution:**
1. **Combine Data:** Pool data from all groups into a single dataset.
2. **Rank Data:** Rank all observations from the combined dataset.
3. **Calculate Rank Sums:** Sum the ranks for each group separately.
4. **Calculate Test Statistic (H):** Use the formula to calculate the Kruskal-Wallis test statistic.
5. **Compare H to Critical Value:** Use tables or software to find the critical value for H.
6. **Determine Significance:** If the calculated H falls above the critical value, reject the null hypothesis.

Performing the Kruskal-Wallis test manually involves ranking the combined data and comparing the test statistic to critical values. Adjust the sample data or significance level as needed for your specific analysis.
</div>

<a id="friedman"></a>

### **Friedman's Test:**

<div align='justify'>

The Friedman test is a non-parametric test used to detect differences in treatments across multiple related groups. It is an extension of the Wilcoxon signed-rank test for more than two related samples. The test assesses whether there are any differences in the distribution of scores among the different groups.

**Counterpart Parametric Test:**
- **Repeated Measures ANOVA:**
  - The Friedman test is the non-parametric equivalent of repeated measures ANOVA. It is employed when the assumptions of repeated measures ANOVA, such as normality and sphericity, are not met.

**Manual Execution:**
1. **Rank Data:** For each group, rank the data across all treatments.
2. **Calculate Friedman Statistic (\(X^2\)):** Use the formula to calculate the Friedman test statistic.
3. **Compare \(X^2\) to Critical Value:** Refer to statistical tables or software to find the critical value for \(X^2\).
4. **Determine Significance:** If \(X^2\) is greater than the critical value, reject the null hypothesis.

Performing the Friedman test manually involves ranking the data and comparing the test statistic to critical values. Adjust the sample data or significance level as needed for your specific analysis.
</div>

<a id="chisquare"></a>

### **Chi-Square Test:**

<div align='justify'>

### Overview:
The Chi-Square test is a non-parametric statistical test used to assess the presence of a significant association between two categorical variables. It is suitable for analyzing contingency tables that display the frequency distribution of two categorical variables.

### Counterpart Parametric Test:
- **Pearson's Chi-Square Test:**
  - The Chi-Square test is the non-parametric equivalent of Pearson's Chi-Square test, utilized when assumptions of the parametric test are not met.

### Manual Execution:
1. **Set Up Hypotheses:**
   - Null Hypothesis $H_0$: There is no significant association between the variables.
   - Alternative Hypothesis $H_a$: There is a significant association between the variables.

2. **Create Contingency Table:**
   - Construct a table with observed frequencies for each combination of categories.

3. **Calculate Expected Frequencies:**
   - For each cell, calculate the expected frequency using: 
   $ \text{Expected Frequency} = \frac{\text{Row Total} \times \text{Column Total}}{\text{Grand Total}} $

4. **Calculate Chi-Square Statistic:**
   - Use the formula: 
   $ \chi = \sum \frac{(O - E)}{E} $
   where $O$ is the observed frequency, $E$ is the expected frequency, and the sum is taken over all cells.

5. **Determine Degrees of Freedom:**
   - Degrees of Freedom $  df = (\text{Number of Rows} - 1) \times (\text{Number of Columns} - 1)$.

6. **Refer to Chi-Square Table:**
   - Compare the calculated Chi-Square statistic with the critical value from the Chi-Square distribution table at the chosen significance level.

7. **Determine Significance:**
   - If the calculated $\chi^2$ is greater than the critical value, reject the null hypothesis, indicating a significant association.

Adjust the significance level as needed for your specific analysis.

</div>

<a id="spearman"></a>

### **Spearman Test:**

<div align='justify'>

The Spearman Rank Correlation test is a non-parametric statistical test used to assess the strength and direction of monotonic relationships between two variables. It calculates the correlation coefficient based on the ranks of the data rather than the raw data values. This makes it robust to outliers and suitable for variables measured on an ordinal or interval scale.

**Counterpart Parametric Test:**
- **Pearson Correlation Coefficient:**
  - The Spearman test is the non-parametric equivalent of the Pearson correlation coefficient. It is employed when the assumptions of the parametric test, such as linearity and normality, are not met.

**Manual Execution:**
1. **Rank Data:**
   - For each variable, rank the data from smallest to largest.
2. **Calculate the Difference in Ranks:**
   - For each pair of ranks, calculate the difference.
3. **Square the Differences:**
   - Square each difference.
4. **Sum the Squares of Differences:**
   - Add up all the squared differences.
5. **Calculate Spearman's Rank Correlation Coefficient ($$):**
   - Use the formula:
   $  = 1 - \frac{6\sum d_i^2}{n(n^2-1)} $
   where $d_i$ is the difference in ranks for each pair, and $n$ is the number of pairs.
6. **Determine Significance:**
   - Compare the calculated $$ to critical values from the Spearman Rank Correlation table at the chosen significance level.

Performing the Spearman Rank Correlation Test manually involves ranking the data, calculating differences, and applying the formula to obtain the correlation coefficient. Adjust the significance level as needed for your specific analysis.
</div>

<a id="onesample"></a>

### **Comparing One-Sample:**

<div align='justify'>
Imagine we want to test whether tests scores of students of a class improved after they have been trained by a new teacher or not, we create two data containig test restuls of students before and after training and perform one-sample test on it.
</div>

```python
import numpy as np
from scipy.stats import wilcoxon

# Generate a fictional dataset
# np.random.seed(42)  # For reproducibility
before_training = [74, 66, 71, 67, 74, 62, 73, 76, 63, 77, 67, 63, 61, 65, 69, 63, 77,
       71, 61, 69]  # Before training scores
after_training = [73, 83, 85, 84, 78, 83, 79, 85, 82, 87, 84, 82, 78, 84, 82, 70, 76,
       78, 70, 81]  # After training scores

# Perform Wilcoxon Signed-Rank Test
statistic, p_value = wilcoxon(before_training, after_training)

# Display the results
print("Wilcoxon Signed-Rank Test:")
print(f"Test Statistic: {statistic}")
print(f"P-value: {p_value}")

# Check for statistical significance
alpha = 0.05
if p_value < alpha:
    print("The difference is statistically significant.")
else:
    print("There is no significant difference.")
```

<a id="twosample"></a>

### **Comparing Two-Samples:**

<div align='justify'>
Now we want to test whether test results of two group of students differ from each other or not, we use two sample test for this part.
</div>

```python
import numpy as np
from scipy.stats import mannwhitneyu

# Generate fictional datasets
np.random.seed(42)  # For reproducibility
group1_scores = np.random.randint(60, 80, 20)  # Scores for Group 1
group2_scores = np.random.randint(70, 90, 20)  # Scores for Group 2

# Perform Mann-Whitney U Test
statistic, p_value = mannwhitneyu(group1_scores, group2_scores)

# Display the results
print("Mann-Whitney U Test:")
print(f"Test Statistic: {statistic}")
print(f"P-value: {p_value}")

# Check for statistical significance
alpha = 0.05
if p_value < alpha:
    print("The difference is statistically significant.")
else:
    print("There is no significant difference.")
```

### **Analysis of Variance:**

<div align='justify'>
Now imagnie we want to test three group of students
</div>

```python
import numpy as np
from scipy.stats import kruskal

# Generate fictional datasets for three groups
np.random.seed(42)
group1_scores = np.random.randint(60, 80, 20)
group2_scores = np.random.randint(70, 90, 20)
group3_scores = np.random.randint(75, 95, 20)

# Perform Kruskal-Wallis test
statistic, p_value = kruskal(group1_scores, group2_scores, group3_scores)

# Display the results
print("Kruskal-Wallis H Test:")
print(f"Test Statistic: {statistic}")
print(f"P-value: {p_value}")

# Check for statistical significance
alpha = 0.05
if p_value < alpha:
    print("There are significant differences between groups.")
else:
    print("There are no significant differences between groups.")
```

<a id="rep-anov"></a>

### **Analysis of Variance Repeated measures:**

<div align='justify'>
In our study, we wanted to investigate the impact of a new teaching method on students' performance. We collected scores from three groups of students before, during, and after the implementation of the new method. To analyze the differences across these time points, we employed the Friedman test, a non-parametric test suitable for repeated measures.
</div>

```python
import numpy as np
from scipy.stats import friedmanchisquare

# Generate fictional repeated measures data
np.random.seed(42)
group1_scores = np.random.randint(60, 80, 10)
group2_scores = np.random.randint(70, 90, 10)
group3_scores = np.random.randint(75, 95, 10)

# Perform Friedman test
statistic, p_value = friedmanchisquare(group1_scores, group2_scores, group3_scores)

# Display the results
print("Friedman Test:")
print(f"Test Statistic: {statistic}")
print(f"P-value: {p_value}")

# Check for statistical significance
alpha = 0.05
if p_value < alpha:
    print("There are significant differences between the repeated measures.")
else:
    print("There are no significant differences between the repeated measures.")
```

<a id="cat"></a>

### **Analysis of Categorical Data:**

<div align='justify'>
In our exploration of student preferences across different classes, we surveyed students from three classes, asking them about their favorite subjects: mathematics, history, and English literature. Now, to understand any potential associations between class and preferred subject, we plan to employ the Chi-Square Test of Independence, a non-parametric test suitable for analyzing categorical data. This will help us determine if there is a significant relationship between the class a student belongs to and their favorite subject, providing valuable insights into student preferences across different classes.
</div>

```python
import numpy as np
from scipy.stats import chi2_contingency

# Generate fictional categorical data
np.random.seed(42)
categories = ["Mathematics", "History", "English Literature"]
observed_data = np.array([
    [30, 20, 10],
    [15, 25, 20],
    [10, 15, 25]
])

# Perform Chi-Square Test of Independence
chi2_stat, p_value, dof, expected = chi2_contingency(observed_data)

# Display the results
print("Chi-Square Test of Independence:")
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Check for statistical significance
alpha = 0.05
if p_value < alpha:
    print("There is a significant association between the categorical variables.")
else:
    print("There is no significant association between the categorical variables.")
```

<a id="correlation"></a>

### **Analysis of Correlation:**

<div align='justify'>

Imagine you're exploring a fascinating world of numbers and observations. You notice that as the number of hours you spend playing your favorite game goes up, your scores tend to change. Intrigued, you wonder if there's a connection between the time spent gaming and your scores. To solve this puzzle, you decide to use a special tool called the Spearman Rank Correlation test.
</div>

```python
import numpy as np
from scipy.stats import spearmanr

# Generate fictional data for two variables
np.random.seed(42)
variable1 = np.random.randint(50, 100, 30)
variable2 = variable1 + np.random.normal(0, 5, 30)

# Perform Spearman Rank Correlation test
correlation_coefficient, p_value = spearmanr(variable1, variable2)

# Display the results
print("Spearman Rank Correlation Test:")
print(f"Correlation Coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")

# Check for statistical significance
alpha = 0.05
if p_value < alpha:
    print("There is a significant correlation between the variables.")
else:
    print("There is no significant correlation between the variables.")
```
